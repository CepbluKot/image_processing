#Задание: создать простой фильтр изображения

В вашем арсенале есть следующие операции:


*   изменение размера изображения
*   перевод в различные цветовые пространства
*   арифметические операции
*   комбинации этих операций
*   и другие, если вы понимаете их смысл

Ниже представлен пример такого фильтра.



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage import io
```

# Домашнее задание 1 - сделать свой крутой фильтр

```python
original_img = io.imread('./tyler.jpg')

plt.title('this is Tyler')
plt.imshow(original_img)
```

```python
# my cool filter 

# инвертировать цвета + трансформировать
# цветовое пространство, тк без этого он будет синим

img_w_filter = original_img

img_w_filter = cv2.bitwise_not(img_w_filter)
img_w_filter = cv2.cvtColor(img_w_filter, cv2.COLOR_BGR2RGB)

plt.title('now Tyler is white')
plt.imshow(img_w_filter)
```

```python
# сложим картинку с фильром и оригинал чтобы 
# выглядело не так топорно

TYLER_1_COEF = .4
TYLER_2_COEF = .9
GAMMA = .0

tyler_1 = original_img
tyler_2 = img_w_filter

tyler_add = cv2.addWeighted(tyler_1, TYLER_1_COEF, tyler_2, TYLER_2_COEF, GAMMA)

plt.title('now Tyler is more white')
plt.imshow(tyler_add)
```

```python
# и превратим картинку в большую картину из майнкрафтра (64 на 64 пикселя) 


tyler_resize = cv2.resize(tyler_add, (64, 64), interpolation=cv2.INTER_LINEAR)


plt.title('now Tyler is a minecraft picture')
plt.imshow(tyler_resize)
```
